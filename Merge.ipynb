## Importando a biblioteca geopandas

```python
import geopandas as gpd
```

## Lendo os arquivos shapefiles com o geopandas

```python
brasil_gdf = gpd.read_file('C:/Users/Tinti/Documents/Codigos Python/Geopandas Youtube/Arquivos/BR_Pais_2022.shp')
usa_gdf = gpd.read_file('C:/Users/Tinti/Documents/Codigos Python/Geopandas Youtube/Arquivos/States_shapefile.shp')
```

## Visualizando a tabela de atributos dos geodataframes

## Visualizando a geometria dos geodataframes

```python
brasil_gdf.plot()
```

```python
usa_gdf.plot()
```

## Igualando o sistema de coordenadas de ambos os geodataframes

```python
brasil_gdf.crs = 'EPSG:4326'
usa_gdf.crs = 'EPSG:4326'
```

## Mesclando ambos os geodataframes

```python
mesclado = gpd.pd.concat([brasil_gdf, usa_gdf])
```
